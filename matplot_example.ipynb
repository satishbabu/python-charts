```python
import matplotlib.pyplot as plt
```

```python
# avoid extra statement to show
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
x
```

```python
# 2 ways of plotting..functional & OOP
# functional way
plt.plot(x,x)
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
# Object Oriented
figure = plt.figure()
axes = figure.add_axes([0.1,0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_title('Title')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
#Subplots
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,x,'r')
axes[1].plot(x,y,'b')
```

```python
#Figure Size
figure = plt.figure()
axes=figure.add_axes([0.1,0.1, 0.8, 0.8])
axes.plot(x,y,color='blue', linewidth=2, marker='o', markersize=10)
```
